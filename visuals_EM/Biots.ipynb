```python
import numpy as np
import itertools
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go
from scipy.integrate import quad

init_notebook_mode(connected=True)

a = 0.5;
shape = "circle" 
perm = 4*np.pi*10e-7 
current = 2.; 
z1 = 1.; 
P = np.array([0,0,1]);
rscal = np.sqrt(a**2+z1**2)
phi_scal = np.pi/2
B = np.empty((0, 3))


def calc_this(shape,phi_scal,z1,a):
    if shape == "circle":
        dI= np.array([-a*np.sin(phi_scal),a*np.cos(phi_scal),0])
        r_vec = np.array([-a*np.cos(phi_scal),-a*np.sin(phi_scal),z1])
        
    return dI,r_vec 

def Biot_equation (current,perm,rscal,dI,r_vec):
    xprod = np.cross(dI,r_vec)
    dB = (current/(4*np.pi*(rscal**3)))*xprod #should multiply current by perm 
    return dB

dI,r_vec = calc_this("circle",phi_scal,z1,a)
dB= 18*Biot_equation(current,perm,r_scal,dI,r_vec) #value of 18 is heuristic for diagram 

def calc_integrand(phi):
    a = 0.5;
    dI= np.array([-a*np.sin(phi),a*np.cos(phi),0])
    r_vec = np.array([-a*np.cos(phi),-a*np.sin(phi),z1])
    result = np.cross(dI,r_vec)
    return result[i]

for j in range(3):
    i=j
    zz = quad(calc_integrand,0,2*np.pi)[0]
    B = np.append(B,zz,axis=None)/2.5 #value of 2.5 is heuristic for diagram 

```

```python
"""Plotting script """


trace1 = go.Scatter3d(x = [a*np.cos(phi_scal),dI[0]+a*np.cos(phi_scal)], 
                      y = [a*np.sin(phi_scal),dI[1]+a*np.sin(phi_scal)], 
                      z = [0,dI[2]],
                      name = "dI", 
                      line = dict(width = 6,color = 'rgb(250,20,0)'),
                      marker = dict(size=[0,15],color= 'rgb(250,20,0)',symbol= 'diamond',opacity=1)
                      
                      )                     


trace2 = go.Scatter3d(x = [0,0+dB[0]], 
                      y = [0,0+dB[1]], 
                      z = [1,1+dB[2]],
                      name = "dB",
                      line = dict(width = 6, color = 'rgb(50,100,200)'),
                      marker = dict(size=[0,15],color= 'rgb(50,100,200)',symbol= 'diamond',opacity=1)
                     );

Btrace = go.Scatter3d(x = [0,0+B[0]], 
                      y = [0,0+B[1]], 
                      z = [1,1+B[2]],
                      name = "B field",
                      line = dict(width = 6, color = 'rgb(255,160,75)'),
                      marker = dict(size=[0,15],color= 'rgb(255,160,122)',symbol= 'diamond',opacity=1)
                     );


trace3 = go.Scatter3d(x = [a*np.cos(phi_scal),r_vec[0]+a*np.cos(phi_scal)],
                      y = [a*np.sin(phi_scal),r_vec[1]+a*np.sin(phi_scal)],
                      z = [0,r_vec[2]],
                      name = "R", 
                      line = dict(width = 6, color = 'rgb(0,220,0)'),
                      marker = dict(size=[0,15],color= 'rgb(0,220,0)',symbol= 'diamond',opacity=1),
                     );


uu = np.linspace(0, 2 * np.pi, 50)
xc = a*np.cos(uu); 
yc = a*np.sin(uu); 
zc = np.zeros_like(xc); 

trace4 = go.Scatter3d( x =  xc ,
                      y = yc,
                      z = zc,
                      name = "circle",
                     mode = "markers",
                     marker = dict(symbol="circle",size=2.5,opacity=1,color= 'rgb(0,0,0)'),
                    );

pointtrace = go.Scatter3d(x = [P[0]],
                          y= [P[1]],
                          z=[P[2]],
                          name = 'Point',
                          marker = dict(size=4, color= 'rgb(214,11,8)')
                         )


data = [trace1,trace2,trace3,trace4,pointtrace,Btrace]

fig = dict(data=data)

iplot(fig)
```
