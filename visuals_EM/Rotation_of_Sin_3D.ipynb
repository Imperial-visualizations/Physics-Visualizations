```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)

import plotly.graph_objs as go
import numpy as np

n = 100
Ex1 = np.linspace(0,10,n)
Ey1 = np.sin(Ex1)
Ez1 = np.zeros(n)

Bx1 = Ex1
By1 = Ez1
Bz1 = Ey1

trace1 = go.Scatter3d(x=Ex1,y=Ey1,z=Ez1, mode='lines', name='Orginal', marker = dict(color='rgb(255,0,0)'))
trace2 = go.Scatter3d(x=Bx1,y=By1,z=Bz1, mode='lines', name='Orginal', showlegend=False, marker = dict(color='rgb(255,0,0)'))

Ecoords1 = [np.array([Ex1[i], Ey1[i], Ez1[i]]) for i in range(0, len(Ex1))]
Bcoords1 = [np.array([Bx1[i], By1[i], Bz1[i]]) for i in range(0, len(Bx1))]

thetax = -np.pi/2.
thetay = -np.pi/4.
thetaz = 3*np.pi/4.
rotmatrixx = np.array([[1, 0, 0],
                    [0, np.cos(thetax), -np.sin(thetax)], 
                    [0, np.sin(thetax),  np.cos(thetax)]])
rotmatrixy = np.array([[np.cos(thetay), 0, np.sin(thetay)],
                    [0, 1, 0], 
                    [-np.sin(thetay), 0,  np.cos(thetay)]])
rotmatrixz = np.array([[np.cos(thetaz), -np.sin(thetaz), 0],
                    [np.sin(thetaz), np.cos(thetaz), 0], 
                    [0, 0, 1]])

Ecoords2 = [np.dot(rotmatrixx,i) for i in Ecoords1]
Ecoords2 = [np.dot(rotmatrixy,i) for i in Ecoords2]
Ecoords2 = [np.dot(rotmatrixz,i) for i in Ecoords2]
Ex2 = [i[0] for i in Ecoords2]
Ey2 = [i[1] for i in Ecoords2]
Ez2 = [i[2] for i in Ecoords2]
trace3 = go.Scatter3d(x=Ex2, y=Ey2, z=Ez2, mode = 'lines', name = 'Rotated', marker = dict(color='rgb(0,0,255)'))

Bcoords2 = [np.dot(rotmatrixx,i) for i in Bcoords1]
Bcoords2 = [np.dot(rotmatrixy,i) for i in Bcoords2]
Bcoords2 = [np.dot(rotmatrixz,i) for i in Bcoords2]
Bx2 = [i[0] for i in Bcoords2]
By2 = [i[1] for i in Bcoords2]
Bz2 = [i[2] for i in Bcoords2]
trace4 = go.Scatter3d(x=Bx2, y=By2, z=Bz2, mode = 'lines', name = 'Rotated', showlegend=False, marker = dict(color='rgb(0,0,255)'))

#Axis Lines
x = np.linspace(-10,10,n)
y = np.linspace(-10,10,n)
z = np.linspace(-10,10,n)
zero = np.zeros(n)

xaxis = go.Scatter3d(x=x, y=zero, z=zero, mode='lines', showlegend=False, marker=dict(color='rgb(0,0,0)'))
yaxis = go.Scatter3d(x=zero, y=y, z=zero, mode='lines', showlegend=False, marker=dict(color='rgb(0,0,0)'))
zaxis = go.Scatter3d(x=zero, y=zero, z=z, mode='lines', showlegend=False, marker=dict(color='rgb(0,0,0)'))

#Arrows
shaft = go.Scatter3d(x=[10,0], y=[0,0], z=[0,0], showlegend=False, mode='lines', line=dict(width=5, color = 'rgb(0,0,0)'))

wing1 = go.Scatter3d(x=[0,1], y=[0,0.5], z=[0,0], showlegend=False, mode='lines', line=dict(width=5, color = 'rgb(0,0,0)'))

wing2 = go.Scatter3d(x=[0,1], y=[0,-0.5], z=[0,0], showlegend=False, mode='lines', line=dict(width=5, color = 'rgb(0,0,0)'))

data=[trace1, trace2, trace3, trace4, xaxis, yaxis, zaxis, shaft, wing1, wing2]

layout = dict(
                width = 700,height=700,
                autosize=True,
                scene = dict(
                aspectmode = 'cube',
                xaxis = dict(range=[-10, 10], autorange=False, zeroline=True),
                yaxis = dict(range=[-10, 10], autorange=False, zeroline=True),
                zaxis = dict(range=[-10, 10], autorange=False, zeroline=True),
                ),
                showlegend = True
                )


fig=go.Figure(data=data,layout=layout)

iplot(fig)
```

## Refactoring

### The rotation matrix

We want to carry out any arbitrary rotation about the origin on a position vector. To do this, we first rotate by $\theta$ (i.e., about the y-axis), then $\phi$ (i.e., about the z-axis). The rotation matrix is thus

$$ \mathbf{A} = \mathbf{A}_z \cdot \mathbf{A}_y =
\begin{pmatrix}
\cos{\phi} & -\sin{\phi} & 0 \\
\sin{\phi} & \cos{\phi} & 0 \\
0 & 0 & 1 \\
\end{pmatrix}
\cdot
\begin{pmatrix}
\cos{\theta} & 0 & \sin{\theta} \\
0 & 1 & 0 \\
-\sin{\theta} & 0 & \cos{\theta} \\
\end{pmatrix}
=
\begin{pmatrix}
\cos{\theta}\cos{\phi} & -\sin{\phi} & \cos{\phi}\sin{\theta} \\
\cos{\theta}\sin{\phi} & \cos{\phi} & \sin{\theta}\sin{\phi} \\
-\sin{\theta} & 0 & \cos{\theta} \\
\end{pmatrix}
$$

### The sine wave

The sine wave is an array of position vectors,

$$\begin{pmatrix}
\mathbf{x}_0 & \mathbf{x}_1 &\mathbf{x}_2 & \ldots \\
\end{pmatrix}$$

which can be represented as

$$\begin{pmatrix}
x^{00} & x^{01} & x^{02} & \ldots \\
x^{10} & x^{11} & x^{12} & \ldots \\
x^{20} & x^{21} & x^{22} & \ldots \\
\end{pmatrix}$$

### The product

So, we're trying to do an element-wise dot-product of the rotation matrix on the array of position vectors

$$A^i_j x^{jk} = y^{ik}$$

To do this with `np.einsum()`, we do

    np.einsum('ij,jk->ik', A, x)

```python
theta = 0.2 * np.pi
phi = 0.5 * np.pi

x_range = np.linspace(0, 10, 100)

sine_data1 = np.array([x_range, np.sin(x_range), np.zeros_like(x_range)])
sine_data2 = np.array([x_range, np.zeros_like(x_range), np.sin(x_range)])

simp_rot = np.array([[np.cos(theta) * np.cos(phi),  -np.sin(phi),  np.cos(phi) * np.sin(theta)],
                     [np.cos(theta) * np.sin(phi),  np.cos(phi),   np.sin(theta) * np.sin(phi)],
                     [-np.sin(theta),               0,             np.cos(theta)              ]])

rot_coords1 = np.einsum("ij,jk->ik", simp_rot, sine_data1)
rot_trace1 = go.Scatter3d(x=rot_coords1[0], y=rot_coords1[1], z=rot_coords1[2],
                         mode = 'lines', name = 'Rotated', marker = dict(color='rgb(0,0,255)'))

rot_coords2 = np.einsum("ij,jk->ik", simp_rot, sine_data2)
rot_trace2 = go.Scatter3d(x=rot_coords2[0], y=rot_coords2[1], z=rot_coords2[2],
                         mode = 'lines', name = 'Rotated', marker = dict(color='rgb(0,255,255)'))

fig=go.Figure(data=[trace1, rot_trace1, rot_trace2], layout=layout)

iplot(fig)
```
