```python
from plotly import __version__
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
from plotly.graph_objs import Scatter, Figure, Layout,Data
import numpy as np
from IPython.display import display, HTML
init_notebook_mode(connected=True)
x = np.linspace(-0.1,2*np.pi+0.1,100)
t= np.linspace(0,4*np.pi,50)
y = np.sin(x)

data=[dict(x=x, y=y,
           name='frame',
           mode='lines', 
           line=dict(width=2, color='blue')),

    ]

layout=dict(width=1000, height=500,
            xaxis=dict(range=[0, 2*np.pi], autorange=False, zeroline=False),
            yaxis=dict(range=[-1.1, 1.1], autorange=False, zeroline=False),
            title='Harmonic Wave in a Vacumm', hovermode='closest',
            updatemenus= [
                {
                    'buttons':[
                        {
                            'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                            'transition': {'duration': 0}}],
                            'label': 'Pause',
                            'method': 'animate'
                        },
                        {
                            'args': [None, {'frame': {'duration': 0, 'redraw': False},
                                'fromcurrent': True, 'transition': {'duration': 0, 'easing': 'quadratic-in-out'}}],
                            'label': 'Play',
                            'method': 'animate'
                        }
                    ]
                }
            ])

frames=[dict(data=[dict(x=x, 
                        y=np.sin(x-t[k])), 

                  ]) for k in range(50)]    
figure = dict(data=data,layout=layout,frames=frames)
iplot(figure)
```
